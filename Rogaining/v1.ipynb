```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import NearestNeighbors
import time

# Generate random number of points between 20 and 30
num_points = np.random.randint(20, 31)

# Generate random x and y coordinates for the points
x_coordinates = np.random.rand(num_points) * 100  # Scale by 100 for better visualization
y_coordinates = np.random.rand(num_points) * 100

# Generate random x and y coordinates for the red point
x_red = np.random.rand() * 100
y_red = np.random.rand() * 100

# Add the red point to the list of coordinates
x_coordinates = np.append(x_coordinates, x_red)
y_coordinates = np.append(y_coordinates, y_red)

# Create a scatter plot for the points
plt.figure(figsize=(8, 6))
plt.scatter(x_coordinates, y_coordinates, color='blue')
plt.scatter(x_red, y_red, color='red')
plt.grid(True)

# Compute nearest neighbors
X = np.column_stack((x_coordinates, y_coordinates))
nbrs = NearestNeighbors(n_neighbors=len(X), algorithm='ball_tree').fit(X)
distances, indices = nbrs.kneighbors(X)

# Draw lines to k nearest neighbors
for j in range(len(X)):
    for i in range(len(indices[j])):
        time.sleep(0.1)  # Introduce a delay of 0.1 seconds
        plt.plot([x_coordinates[j], x_coordinates[indices[j][i]]], [y_coordinates[j], y_coordinates[indices[j][i]]], color='black')
        plt.draw()
        plt.pause(0.001)

plt.show()
```
